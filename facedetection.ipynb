```python
import cv2
import dlib
import numpy as np 
```

```python
#capture the frames from my primary camera
cap = cv2.VideoCapture(0) #0 indicates primary camera

#face detector model
face_detector = dlib.get_frontal_face_detector()

#capture frames continuosly
while True: 
    ret, frame = cap.read() #ret: boolean value: 1 if its able to read it, frame = actual fram which gets captured
    frame = cv2.flip(frame, 1) #1- vertically flips the image

    #RGB to grayscale :grayscaling done so that lot of noise can be reduced in image, less in size and easier and faster to process the specific image
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_detector(gray) 
    
    #iterate through all the faces and draw rectangle around each face and also number it
    i=0
    for face in faces:
        #Find the coordinates of the face
        x, y= face.left(), face.top()
        x1, y1 = face.right(), face.bottom()

        #Draw the rectangle
        cv2.rectangle(frame,(x,y),(x1,y1),(0,255,0), 2) #coordinates - color - thickness
        
        i = i+1
                                        #face num, text coordinate, text font, text color, aspect ratio, its thickness
        cv2.putText(frame, "Face Num" + str(i), (x-10,y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)

    cv2.imshow('frame', frame)

    #Code to come out of infinite loop / interrupt the execution

    if(cv2.waitKey(1) & 0xFF == ord('q')):
        break

cap.release()
cv2.destroyAllWindows()
```
